```python
import seaborn as sns            
import matplotlib.pyplot as plt
import pandas as pd
```

#### Import data

```python
path = '/home/victhor02/Desktop/doc/3B/apa/hearth-disease-predictor/data/sheart_train_ok.csv'
data = pd.read_csv(path)
```

### Exploratory data analysis

```python
for idx, feature in enumerate(list(data.columns)[:-1]):
  fg = sns.FacetGrid(data, col='chd', height=2)
  fg.map(sns.histplot, feature,kde=True).add_legend()
plt.show()
```

```python
for idx, feature in enumerate(list(data.columns)[:-1]):
    sns.histplot(data=data, x=feature, hue="chd",kde=True, bins=30, multiple="stack")
    plt.show()
```

```python
sns.pairplot(data, hue="chd")
plt.show()
```

```python
# Crear el gràfic de heatmap
fig, ax = plt.subplots(figsize=(8, 8))
heatmap = sns.heatmap(data.corr(),
                      annot=True,
                      linewidths=1,
                      linecolor='black',
                      cmap='coolwarm',
                      cbar=True,
                      square=True,
                      ax=ax)

# Personalitzar les etiquetes dels eixos
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

# Afegir títol
plt.title("Correlació entre les variables", fontsize=16)

# Mostrar el gràfic
plt.show()
```
